# Word Embeddings 

Converting words to vectors (vector embeddings)



Deep learning Trained Model for vector embeddings

## Word2Vec

uses a neural network model to learn a word association from large corpus of text. can detect synonyms and additional words for partial sentences. developed by Google.

**Feature representation**



### Cosine Similarity 

Angle between two vectors cos(theta)

Distance = 1 - cosine similarity 





## CBOW [Continous Bag of Words]

fully connected nerual network

Window Size = 5 (example) --> middle layer

feature representation = window size only (5 here)




                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                           

## Skipgram




if dataset is small --> CBOW 
if dataset is huge ---> skipgram 

Improve CBOW or skipgram: 
1) Increase dataset for training 
2) increase the window size - increased vector dimension 

Word2Vec has 3 billion words and feature representation of 300 dimension vectors
